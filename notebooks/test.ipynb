```python
from regulatory_rnn import RegulatoryRNN, GapGeneDataset, AdaptPulseDataset
```

```python
dataset = GapGeneDataset()
model = RegulatoryRNN(
    n_targets=dataset.n_targets,
    n_stimuli=dataset.n_stimuli,
    n_observed_targets=dataset.n_observed_targets,
    n_networks=dataset.n_targets,
    stepsize=0.2
)
model.fit(dataset)
network = model.infer_network(dataset, refit=False)
network.draw()
```
